```python
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

fig,ax = plt.subplots(figsize =(8,7))
fig.set_facecolor('#b2b2b2') #cambia el fondo del gráfico, por defecto es blanco, se puede usar directamente grey
# leer más en https://taniaizquierdo.com/color-hexadecimal/#:~:text=El%20color%20hexadecimal%20es%20un,por%20sus%20siglas%20en%20ingl%C3%A9s).

"""
#b2b2b2 es un valor hexadecimal que representa un color en el modelo de colores RGB (rojo, verde y azul)
utilizado en la programación y el diseño. Los valores hexadecimales se utilizan para representar los valores de los componentes rojo, verde y azul
(en ese orden) en un solo número de 6 dígitos.

En este caso, #b2b2b2 representa un gris medio, donde el componente rojo tiene un valor de b2 en hexadecimal (178 en decimal),
el componente verde tiene un valor de b2 en hexadecimal (178 en decimal) y el componente azul tiene un valor de b2 en hexadecimal (178 en decimal).

Este valor hexadecimal se puede utilizar en diferentes contextos en los que se necesite especificar un color,
como en el fondo de una figura o en el color de una línea o una barra en un gráfico.

"""

ax.pie(data, labels = cars,radius=0.7, startangle=90, explode= (0,0,0,0,0,0.1), autopct= "%1.2f%%", colors=sns.color_palette("hls",7))

"""
labels: lista con los nombres de las etiquetas que se mostrarán en el gráfico.
radius: radio del gráfico circular.
startangle: ángulo inicial de la primera división del gráfico.
explode: lista que especifica la separación de cada división del gráfico circular respecto al centro.
autopct="%1.1f%%" significa que se mostrará el valor porcentual con un decimal y un signo de porcentaje
colors: es una variable que se utiliza para almacenar una paleta de colores generada con la función color_palette() del paquete seaborn en Python.
"""
plt.legend(loc="lower left",labels=cars,fontsize=10)
plt.show()
```

# Grafico 1: histograma para ver distribucion del salario

```python
# Orientado a objetos
mpl.style.use('classic')
"""
establece el estilo de visualización predeterminado de Matplotlib como 'classic'.
 Matplotlib como vimos en clase tiene varios estilos de visualización incorporados, como 'ggplot', 'bmh', etc.
"""
fig, ax = plt.subplots(figsize=(10,6))
"""
Esta línea crea una figura y un conjunto de ejes en los que se trazará el histograma.
La figura es el lienzo en el que se dibuja la visualización, mientras que los ejes son los objetos que se utilizan para
trazar los datos en la figura. Se establece el tamaño de la figura en 10 pulgadas de ancho y 6 pulgadas de alto
"""
ax.hist(x= data.Salary, bins=None, orientation='vertical')
"""
Esta línea traza el histograma en los ejes creados anteriormente.
La función hist() recibe varios argumentos, como los datos que se van a trazar
(en este caso, los salarios del conjunto de datos 'data'), el número de bins
(si no se especifica, Matplotlib calculará automáticamente el número óptimo de bins),
y la orientación del histograma ('vertical' o 'horizontal').
"""
ax.set_xlabel('Salario')
ax.set_ylabel('Frecuencia')
ax.set_title('Histograma de Salario')

"""
Estas líneas hacen referencia a las etiquetas
"""
```

```python
# Orientado a estados

plt.figure(figsize=(10,6))
plt.hist(x= data.Salary, bins=None, orientation='vertical')
plt.xlabel('Salario')
plt.ylabel('Frecuencia')
plt.title('Histograma de Salario')
```

Se logra ver que la distribucion de los datos es asimetrica a derecha con la mayoria de valores en la parte inferior de la distribucion (valores tipicos entre 50000-100000). Se logra ver tambien que hay presencia de valores por encima de 2000000

```python
data_2=data.groupby('Department')[['EmpID']].count()
data_2
```

```python
plt.figure(figsize=(10,6))
plt.pie(data_2.EmpID, labels = data_2.index,autopct='%1.1f%%')
plt.title('Piechart de distribucion de departamento')

#una versión no recomendada
```

```python
fig,ax = plt.subplots(figsize =(10, 10))
ax.pie(data_2.EmpID,autopct='%1.1f%%', pctdistance = 1.1)
plt.legend(loc="lower right",labels=data_2.index,fontsize=10)
ax.set_title('Piechart de distribucion de departamento')

#Ten en cuenta que también puedes personalizar la distancia de estas etiquetas desde el origen con pctdistance
#y mostrarlas en lugar de las etiquetas de los grupos. El valor por defecto del argumento es 0.6.

# revisar siempre la documentación oficial https://python-charts.com/es/parte-todo/pie-chart-matplotlib/
```
