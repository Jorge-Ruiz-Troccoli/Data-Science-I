## Texto a audio

```python
!pip install gTTS

"""
La librería GTTS (Google Text-to-Speech) es una herramienta de Python que permite convertir texto en voz utilizando los
servicios de síntesis de voz de Google.
"""
```

```python
from gtts import gTTS
from IPython.display import Audio
tts = gTTS('hola Jorge, cómo estás?',lang='es')  # acá se introduce el texto para pasar a audio
tts.save('ejemplo_saludo.mp3') #se guarda el audio
sound_file = 'ejemplo_saludo.mp3'
Audio(sound_file, autoplay=True) #podemos reproducirlo
```

```python
from google.colab import files
files.download('ejemplo_saludo.mp3')

#además se puede descargar el audio a nuestro entorno local
```

```python
from gtts import gTTS
from IPython.display import Audio
tts = gTTS('hello jorge how are you?',lang='en') #es lo mismo que antes pero ahora en ingles
tts.save('1.wav')
sound_file = '1.wav'
Audio(sound_file, autoplay=True)

#revisar la documentación en https://gtts.readthedocs.io/en/latest/index.html
```

```python
from deep_translator import GoogleTranslator

# Instanciar el traductor de Google
translator = GoogleTranslator(source='auto', target='en')

# Texto a traducir
text = "hola mundo"

# Traducir el texto
translated_text = translator.translate(text)

# Imprimir el resultado
print(f"Texto original: {text}")
print(f"Traducción (english): {translated_text}")
```
